1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Understanding Metrics in Real-World Applications
In machine learning, understanding when and why to use specific metrics—Accuracy, Sensitivity, Specificity, and Precision—is crucial because each measures different aspects of model performance and is relevant depending on the specific problem context.


When It's Relevant:

Balanced Datasets: Accuracy is suitable when the dataset has roughly equal numbers of instances for each class and when false positives and false negatives have similar costs.
Example: Weather prediction models (e.g., predicting rain vs. no rain) where both types of errors are equally critical.
Why It Matters: Provides an overall snapshot of model performance but can be misleading when datasets are imbalanced.

When It's Relevant:

Critical for Missing Positive Cases: Sensitivity is crucial when failing to detect positives (false negatives) has severe consequences.
Example: Disease diagnosis, such as cancer screening, where missing a positive case (false negative) could be life-threatening.
Why It Matters: High sensitivity minimizes false negatives, ensuring that most positive cases are detected.

When It's Relevant:

Critical for Avoiding False Alarms: Specificity is essential when false positives lead to unnecessary costs or interventions.
Example: Fraud detection in banking, where wrongly flagging legitimate transactions can inconvenience customers.
Why It Matters: High specificity ensures that false positives are minimized, reducing unnecessary actions or disruptions.

When It's Relevant:

Critical in Scenarios with High Costs of False Positives: Precision matters when positive predictions need to be highly reliable.
Example: Spam detection in email systems, where high precision ensures that flagged spam emails are genuinely spam, avoiding the risk of misclassifying important emails.
Why It Matters: High precision minimizes false positives, ensuring positive predictions are trustworthy.

Conclusion
The choice of metric depends on the problem context:

Use accuracy for balanced datasets where all types of errors are equally costly.
Use sensitivity when missing a positive case is more harmful than a false alarm.
Use specificity when avoiding false positives is critical.
Use precision when you care about the reliability of positive predictions.
This nuanced understanding helps select the right metric for evaluating model performance effectively.

1. Create an 80/20 Train-Test Split
To split the dataset into 80% training and 20% testing data, we use the train_test_split() method from Scikit-learn:

````python
import pandas as pd
from sklearn.model_selection import train_test_split

# 加载数据
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# 数据预处理
ab_reduced = ab.drop(columns=['Weight_oz', 'Width', 'Height'])  # 删除无关列
ab_reduced_noNaN = ab_reduced.dropna()  # 删除包含缺失值的行

# 确认数据集准备就绪
print("Preprocessed dataset:")
print(ab_reduced_noNaN.head())

# 分割训练集和测试集
train_set, test_set = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# 打印训练集和测试集大小
print(f"Training set size: {len(train_set)}")
print(f"Testing set size: {len(test_set)}")
````

2. Train a DecisionTreeClassifier
We train a decision tree classifier using List Price as the feature and Hard_or_Paper as the target. The tree has a maximum depth of 2.

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Define features and target
X_train = train_set[['List Price']]
y_train = train_set['Hard_or_Paper']

# Encode the target variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Train the DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train_encoded)
````

3. Visualize the Fitted Decision Tree
The fitted tree is visualized using tree.plot_tree() to show decision rules, class probabilities, and predictions.

````python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(12, 6))
plot_tree(clf, feature_names=['List Price'], class_names=label_encoder.classes_, filled=True)
plt.show()
````

Sensitivity = TP / (TP + FN)
Specificity = TN / (TN + FP)

````python
# Define the calculate_metrics function
def calculate_metrics(conf_matrix):
    TN, FP, FN, TP = conf_matrix.ravel()
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    return accuracy, sensitivity, specificity

# Check features used during training
print("Training features:", clf2.feature_names_in_)

# Align test dataset features
X_test = X_test[clf2.feature_names_in_]

# Run predictions
clf2_predictions = clf2.predict(X_test)

# Confusion matrix and metrics
conf_matrix_clf2 = confusion_matrix(y_test_encoded, clf2_predictions)
accuracy_clf2, sensitivity_clf2, specificity_clf2 = calculate_metrics(conf_matrix_clf2)

# Print results
print(f"clf2 - Accuracy: {accuracy_clf2}, Sensitivity: {sensitivity_clf2}, Specificity: {specificity_clf2}")
````

Key Insights
Accuracy measures overall correctness.
Sensitivity ensures fewer false negatives (important for identifying positives like "Hardcover").
Specificity reduces false positives (important for avoiding incorrect "Paperback" predictions).

### Differences Between Coefficients in Linear Regression and Feature Importances in Decision Trees

1. **Nature of Relationships**  
   - **Linear Regression**: Assumes a linear and additive relationship between predictors and the target variable. Coefficients represent the direct effect of predictors on the outcome.  
   - **Decision Trees**: Capture non-linear and hierarchical relationships. Feature importances reflect the overall contribution of predictors to model performance, not direct effects.

2. **Interaction Handling**  
   - **Linear Regression**: Does not account for interactions unless explicitly included as interaction terms.  
   - **Decision Trees**: Naturally capture interactions through hierarchical splits, allowing predictors to influence outcomes jointly.

3. **Interpretability**  
   - **Linear Regression**: Coefficients are easily interpretable as the effect size of a predictor on the outcome, assuming all other variables are constant.  
   - **Decision Trees**: Feature importances are relative and indicate which predictors are most influential, but do not provide specific quantitative relationships.

4. **Sensitivity to Scale**  
   - **Linear Regression**: Coefficients depend on the scale of predictors. Standardizing variables is necessary for comparing their influence.  
   - **Decision Trees**: Feature importance is scale-invariant, as splits are based on value ordering rather than absolute magnitude.

5. **Predictor Influence**  
   - **Linear Regression**: Predictors have a fixed effect on the target variable, independent of other predictors.  
   - **Decision Trees**: The effect of a predictor depends on the splits and paths in the tree, making it context-dependent.

### Summary
- Linear regression provides direct and interpretable coefficients for linear relationships but requires careful handling of interactions and scaling.  
- Decision trees offer flexibility in capturing non-linear relationships and interactions, but their feature importances are relative and less direct to interpret.  

These differences make linear regression more suitable for problems requiring straightforward interpretability, while decision trees are better for capturing complex patterns and interactions in data.
